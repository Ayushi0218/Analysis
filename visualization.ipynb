```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly. express as px
```

```python
penguins= sns.load_dataset("penguins")
penguins
```

```python
penguins['bill_depth_mm'].plot()
```

```python
penguins['bill_depth_mm'].plot(linestyle='',
                               marker='s',
                               markersize=5,
                               alpha=0.5,
                               figsize=(5,5),
                               markerfacecolor='red',
                               markeredgecolor='black',)
```

```python
penguins['bill_depth_mm'][:50].plot(kind='bar',figsize=(15,4),color='green')
```

```python
penguins['bill_depth_mm'][:50].plot(kind='area',figsize=(15,4),color='green')
```

```python
print(sns.get_dataset_names())
```

```python
penguins.plot(kind='scatter',x='bill_length_mm',y='bill_depth_mm',s=100,alpha=.5)
```

```python
penguins.plot(kind='scatter',x='bill_length_mm',y='flipper_length_mm',s=100,alpha=.5)
```

```python
fig,axs=plt.subplots(1,2, figsize =(15,4),constrained_layout=True)
penguins.plot(kind='scatter',x='bill_length_mm',y='bill_depth_mm',s=100,alpha=.5,ax=axs[0],title='penguin bill data',color='red')
penguins.plot(kind='scatter',x='bill_length_mm',y='flipper_length_mm',s=100,alpha=.5,ax=axs[1],title='penguin bill data')
plt.show()
```

```python
penguins['species'].value_counts().plot(kind='pie',
                                       autopct='%1.1f%%', # add in percentages
                                       figsize=(3,3),
                                       title='Species',
                                       ylabel='',
                                       wedgeprops=dict(
                                        width=0.5,
                                       ))
```

```python
tips = sns.load_dataset("tips")
sns.relplot(data=tips, x="total_bill", y="tip")
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker")
```

```python
sns.relplot(
    data=tips,
    x="total_bill", y="tip", hue="smoker", style="smoker"
)
```

```python
sns.relplot(
    data=tips,
    x="total_bill", y="tip", hue="smoker", style="time",
)
```

```python
sns.relplot(
    data=tips, x="total_bill", y="tip", hue="size",
)
```

```python
dowjones = sns.load_dataset("dowjones")
sns.relplot(data=dowjones, x="Date", y="Price", kind="line")
```

```python
fmri = sns.load_dataset("fmri")
sns.relplot(data=fmri, x="timepoint", y="signal", kind="line")
```

```python
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", errorbar="sd",
)
```

```python
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal",
    estimator=None,
)
```

```python
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", hue="region", style="event",
    dashes=False, markers=True,
)
```

```python
sns.relplot(
    data=fmri.query("event == 'stim'"), kind="line",
    x="timepoint", y="signal", hue="region",
    units="subject", estimator=None,
)
```
